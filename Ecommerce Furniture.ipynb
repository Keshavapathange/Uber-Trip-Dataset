```python
#LINE CHART
```

```python
sales = [120, 150, 180, 170, 200]
months = ["Jan", "Feb", "Mar", "Apr", "May"]

plt.figure()
plt.plot(months, sales)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
#PIE CHART
```

```python
categories = ["Electronics", "Furniture", "Clothing"]
share = [40, 35, 25]

plt.figure()
plt.pie(share, labels=categories, autopct="%1.1f%%")
plt.title("Category Share")
plt.show()
```

```python
#BAR GRAPH
```

```python
products = ["Chair", "Table", "Sofa", "Bed"]
revenue = [500, 700, 1200, 900]

plt.figure()
plt.bar(products, revenue)
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.title("Revenue by Product")
plt.show()
```
